#Q1 Explaination
Given:

Matrix1: order m x k
Matrix2: order k x n
Matrix3=Matrix1 x Matrix2

Since we mutiply k elements of every rows of matrix1 with k elements of every corresponding columns of matrix2. For multiplication of one row with one corresponding column there are k multiplications. And now we have to add these newly obtained k values, so there will be k-1 additions. We did this process once, but we have to do this m x n times as Matrix3 has m x n elements. So there are k x m x n multiplications and (k-1) x m x n additions and total (2k-1) x m x n additions and multiplications.

```python
#Q1
m=int(input())
n=int(input())
k=int(input())
print('multiplications=',k*m*n)
print('multiplications=',(k-1)*m*n)
Answer=(2*k-1)*m*n
print('Total additions and multiplications=',Answer)
```

In #Q3 I used quickselect function to partition the list and then find the median

```python
#Q3
#M1:

A=list(map(int,input().split()))
def M1(A):
  max=A[0]
  for i in range((len(A)//2)+1):

    for j in range(len(A)-i):
      if A[j]>max:
       max=A[j]
    A[j],A[-i]=A[-1],A[j]


def partition(A,start,end):
  i=start+1
  pivot=A[start]
  j=end

  while i<=end and A[i]<pivot:
    i+=1
  while j>=start and A[j]>pivot:
    j-=1
  if i<=j:
    A[i],A[j]=A[j],A[i]
    i+=1
    j-=1

  if i<=end:
    A[i-1],A[start]=A[start],A[i-1]
    return i-1
  else:
    A[end],A[start]=A[start],A[end]
    return end

#M2

def quickselect1(A,start,end,k):
  if start>=end:

    return A[start]
  index=partition(A,start,end)
  if start+k-1==index:
    return A[index+1]
  elif start+k-1<index:
    return quickselect1(A,start,index-1,k)
  elif start+k-1>index:
    return quickselect1(A,index+1,end,k-(index+1-start))


def quickselect2(A,start,end,k1,c):
  if start>=end:
    return A[start]
  index=partition(A,start,end)

  if index==start+k1-1:

    c+=1
    if c==1:
      v1=A[index]
    if c==2:
      v2= A[index]
      return (v1+v2)/2

    return quickselect2(A,index+1,end,k1+1-(index+1-start))

  if start+k1-1<index:
    return quickselect2(A,start,index-1,k)
  elif start+k1-1>index:
    return quickselect2(A,index+1,end,k-(index+1-start))

if len(A)%2==0:
  print(quickselect2(A,0,len(A)-1,len(A)//2),0)
else:
  print(quickselect1(A,0,len(A)-1,(len(A))//2))
```

```python
#Q5

import jax.numpy as jnp
from jax import grad

def check(inputs):
    x, y = inputs
    return (x**2)*y + (y**3)*(jnp.sin(x))


gradient = grad(check)


point = jnp.array([10.0, 1.0])
print(gradient(point))
```
